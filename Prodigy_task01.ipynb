Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style
```

```python
df = pd.read_csv('/content/testdata2.csv')
df.head()
```

```python
df_final = df.drop(['CCA3', 'Rank',  'Capital', 'Continent'], axis=1)
```

**BAR PLOT**

```python
chosen_country = 'Andorra'

#Initialize variables
country_data = df_final[df_final['Country/Territory'] == chosen_country]

years = country_data.columns[2:].astype(int)
population = country_data.iloc[:, 2:].values.flatten()

#Create a bar plot
style.use('ggplot')
plt.figure(figsize=(5, 3))
color=['royalblue','red','purple','yellow']
plt.bar(years, population, color=color, width=0.5)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')

#Display the plot
plt.tight_layout()
plt.show()
```

**HISTOGRAM**

```python
#Create a histogram
plt.hist(population, bins=20, color='limegreen', edgecolor='black')

#Display the histogram
plt.tight_layout()
plt.show()
```
